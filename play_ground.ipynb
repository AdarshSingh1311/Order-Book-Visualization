```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
with open('price_qoute.pickle1' , 'rb') as pickle_file :
    price_qoute_dct = pickle.load(pickle_file)

price_df = price_qoute_dct['price']    
qoute_df = price_qoute_dct['qoute']
```

```python
price_df.reset_index(drop = True , inplace= True)
price_df['last'].plot()
```

```python
qoute_df
```

```python
qoute_df.reset_index(drop = True , inplace= True)
plt.plot( qoute_df['BestBuyRate-1'] )
plt.plot( qoute_df['BestBuyRate-2'] )
plt.plot( qoute_df['BestSellRate-1'] )
plt.plot( qoute_df['BestSellRate-2'] )
```

```python
price_qoute_df = price_df.join(qoute_df.set_index('time') , on = 'ltt' )
price_qoute_df
```

```python

plt.figure(figsize = (12,6))

plt.plot(price_qoute_df['last'])

plt.plot( price_qoute_df['BestBuyRate-1'] , color = 'green' )
plt.plot( price_qoute_df['BestBuyRate-2'] ,   color = 'green' )
plt.plot( price_qoute_df['BestBuyRate-3'] ,  color = 'green' )
plt.plot( price_qoute_df['BestBuyRate-4']  , color = 'green' )
plt.plot( price_qoute_df['BestBuyRate-5']  , color = 'green')

plt.plot( price_qoute_df['BestSellRate-1'] , color = 'red' )
plt.plot( price_qoute_df['BestSellRate-2'] , color = 'red')
plt.plot( price_qoute_df['BestSellRate-3'] , color = 'red')
plt.plot( price_qoute_df['BestSellRate-4'] , color = 'red')
plt.plot( price_qoute_df['BestSellRate-5'] , color = 'red')
```

```python
tick_size = 0.05   # nifty fut tick size
tick_size_plt = 0.05*5

bid = []
for depth_no in np.arange(1,6) :
    col = price_qoute_df[f'BestBuyQty-{depth_no}']
    price_qoute_df[f'BestBuyQty-{depth_no}'] = col*tick_size_plt / col.max()
  
    bid += zip( price_qoute_df.index,
                price_qoute_df[f'BestBuyRate-{depth_no}'], 
                price_qoute_df[f'BestBuyQty-{depth_no}']  
                    )
    

ask = []
for depth_no in np.arange(1,6) :
    col = price_qoute_df[f'BestSellQty-{depth_no}']
    price_qoute_df[f'BestSellQty-{depth_no}'] = col*tick_size_plt / col.max()

    ask += zip( price_qoute_df.index,
                price_qoute_df[f'BestSellRate-{depth_no}'], 
                price_qoute_df[f'BestSellQty-{depth_no}']  
                    )
    
                                           
plt.figure(figsize = (12,6))

for indx,bid_price,bid_qty in bid :
    plt.bar( indx,-bid_qty, bottom=bid_price , width = 1 ,color = 'green' , alpha = 1)


for indx,ask_price,ask_qty in ask :
    plt.bar( indx, ask_qty, bottom=ask_price , width = 1 ,color = 'red' , alpha = 1 )  


price_qoute_df['bid_ask_spread'] = price_qoute_df['BestSellRate-1'] - price_qoute_df['BestBuyRate-1']   

spread_zip = zip( price_qoute_df.index ,
                  price_qoute_df['bid_ask_spread'],
                  price_qoute_df['BestSellRate-1'] 
                      ) 

price_qoute_df['sell_price'] = np.where( abs(price_qoute_df['BestSellQty-1'] - price_qoute_df['last']) > 
                                         abs(price_qoute_df['BestBuyQty-1'] - price_qoute_df['last']), price_qoute_df['last'], np.nan )

price_qoute_df['buy_price'] = np.where( abs(price_qoute_df['BestSellQty-1'] - price_qoute_df['last']) < 
                                         abs(price_qoute_df['BestBuyQty-1'] - price_qoute_df['last']), price_qoute_df['last'], np.nan )

for indx,spread,ask_price in spread_zip  :
    plt.bar(indx,-spread,bottom = ask_price , width = 1 , color = 'lemonchiffon' ,alpha = 1 )

plt.plot(price_qoute_df['last'].astype(float) , color = 'blue')  

plt.scatter(price_qoute_df.index , price_qoute_df.buy_price , edgecolors= 'black' , color = 'green', linewidths = 1)
plt.scatter(price_qoute_df.index , price_qoute_df.sell_price , edgecolors = 'black', color = 'red' , linewidths= 1)

plt.title('ORDER FLOW')
plt.xlabel('TIME')
plt.ylabel('PRICE')

plt.legend(['tick' , 'bid' , 'ask'])
```
